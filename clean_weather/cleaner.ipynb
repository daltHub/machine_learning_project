```python
import numpy as np
import datetime
import pandas as pd
```

```python
weather = pd.read_csv('weather.csv')
weather = weather.drop(axis= 1, columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
availability = pd.read_csv('weather_availability.csv')
weather
```

```python
m1 = weather[weather.station_id == 'M1']
m1_avail = availability[weather.station_id == 'M1']
m1.to_csv('output_data/m1_full.csv')
m1_avail.to_csv('output_data/m1_avail_full.csv')

m2 = weather[weather.station_id == 'M2']
m2_avail = availability[weather.station_id == 'M2']
m2.to_csv('output_data/m2_full.csv')

m3 = weather[weather.station_id == 'M3']
m3_avail = availability[weather.station_id == 'M3']
m3.to_csv('output_data/m3_full.csv')

m4 = weather[weather.station_id == 'M4']
m4_avail = availability[weather.station_id == 'M4']
m4.to_csv('output_data/m4_full.csv')

m5 = weather[weather.station_id == 'M5']
m5_avail = availability[weather.station_id == 'M5']
m5.to_csv('output_data/m5_full.csv')

m6 = weather[weather.station_id == 'M6']
m6_avail = availability[weather.station_id == 'M6']
m6.to_csv('output_data/m6_full.csv')

Belmullet = weather[weather.station_id == 'Belmullet-AMETS']
Belmullet_avail = availability[weather.station_id == 'Belmullet-AMETS']
Belmullet.to_csv('output_data/Bellmullet_full.csv')

fs1 = weather[weather.station_id == 'FS1']
fs1_avail = availability[weather.station_id == 'FS1']
fs1.to_csv('output_data/fs1_full.csv')


```

```python
# Remove unused columns, add time and date seperately
# m1.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
m1['date'] = pd.to_datetime(m1['time']).dt.date
m1['time'] = pd.to_datetime(m1['time']).dt.time
m1.reset_index()
m1.to_csv('output_data/m1_full_columns_removed.csv')

# m2.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
m2['date'] = pd.to_datetime(m2['time']).dt.date
m2['time'] = pd.to_datetime(m2['time']).dt.time
m2.reset_index()
m2.to_csv('output_data/m2_full_columns_removed.csv')

# m3.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
m3['date'] = pd.to_datetime(m3['time']).dt.date
m3['time'] = pd.to_datetime(m3['time']).dt.time
m3.reset_index()
m3.to_csv('output_data/m3_full_columns_removed.csv')

# m4.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
m4['date'] = pd.to_datetime(m4['time']).dt.date
m4['time'] = pd.to_datetime(m4['time']).dt.time
m4.reset_index()
m4.to_csv('output_data/m4_full_columns_removed.csv')

# m5.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
m5['date'] = pd.to_datetime(m5['time']).dt.date
m5['time'] = pd.to_datetime(m5['time']).dt.time
m5.reset_index()
m5.to_csv('output_data/m5_full_columns_removed.csv')

# m6.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
m6['date'] = pd.to_datetime(m6['time']).dt.date
m6['time'] = pd.to_datetime(m6['time']).dt.time
m6.reset_index()
m6.to_csv('output_data/m6_full_columns_removed.csv')

# Belmullet.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
Belmullet['date'] = pd.to_datetime(Belmullet['time']).dt.date
Belmullet['time'] = pd.to_datetime(Belmullet['time']).dt.time
Belmullet.reset_index()
Belmullet.to_csv('output_data/Belmullet_full_columns_removed.csv')

# fs1.drop(columns=['RelativeHumidity', 'DewPoint', 'AirTemperature', 'QC_Flag'])
fs1['date'] = pd.to_datetime(fs1['time']).dt.date
fs1['time'] = pd.to_datetime(fs1['time']).dt.time
fs1.reset_index()
fs1.to_csv('output_data/fs1_full_columns_removed.csv')
```

```python
# Now remove rows with incomplete data.
m1_valid = m1[ m1.WaveHeight >0][m1.WindSpeed>0][m1.AtmosphericPressure>0]
m1_valid.to_csv('valid_data/m1_valid.csv')

m2_valid = m2[ m2.WaveHeight >0][m2.WindSpeed>0][m2.AtmosphericPressure>0]
m2_valid.to_csv('valid_data/m2_valid.csv')

m3_valid = m3[ m3.WaveHeight >0][m3.WindSpeed>0][m3.AtmosphericPressure>0]
m3_valid.to_csv('valid_data/m3_valid.csv')

m4_valid = m4[ m4.WaveHeight >0][m4.WindSpeed>0][m4.AtmosphericPressure>0]
m4_valid.to_csv('valid_data/m4_valid.csv')

m5_valid = m5[ m5.WaveHeight >0][m5.WindSpeed>0][m5.AtmosphericPressure>0]
m5_valid.to_csv('valid_data/m5_valid.csv')

m6_valid = m6[ m6.WaveHeight >0][m6.WindSpeed>0][m6.AtmosphericPressure>0]
m6_valid.to_csv('valid_data/m6_valid.csv')

Bellmullet_valid = Belmullet[ Belmullet.WaveHeight >0][Belmullet.WindSpeed>0][Belmullet.AtmosphericPressure>0]
Bellmullet_valid.to_csv('valid_data/Bellmullet_valid.csv')

fs1_valid = fs1[ fs1.WaveHeight >0][fs1.WindSpeed>0][fs1.AtmosphericPressure>0]
fs1_valid.to_csv('valid_data/fs1_valid.csv')
```
