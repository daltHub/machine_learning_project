```python
import numpy as np
import datetime
import pandas as pd
```

```python
# locations of data files
wave_spectral_df = pd.read_csv('cleaner_data/wave_spectral.csv')
wave_spectral_availability = pd.read_csv('cleaner_data/wave_spectral_availability.csv')
print(wave_spectral_df)
```

```python
# Query dataframe for availability
dates_to_delete = wave_spectral_availability.query(expr='PeakPeriod != 100 and PeakDirection != 100 and PeakSpread !=100 and SignificantWaveHeight != 100 and EnergyPeriod !=100 and MeanWavePeriod_Tm01 !=100 and MeanWavePeriod_Tm02 !=100')['Date']

dates_to_delete = dates_to_delete.drop_duplicates()
dates_to_delete
```

```python
wave_spectral_df['Date'] = pd.to_datetime(wave_spectral_df['time']).dt.date
wave_spectral_df['Time'] = pd.to_datetime(wave_spectral_df['time']).dt.time
wave_spectral_df
```

```python
this_date = 'Clew Bay'
# x = wave_spectral_df.query('station_id == @this_date')
wave_spectral_df.drop(wave_spectral_df.index[wave_spectral_df['station_id'] == this_date], inplace = True)
# only_valid.drop(only_valid.index[only_valid['Date'] == invalid_date], inplace = True)
wave_spectral_df
```
